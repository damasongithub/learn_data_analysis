# Data Clearing

## Handling Missing Data with pandas

we use the .notnull() method or the .isnull() method to check whether there is or there is no null values in the data. 

```python
# we can start with an info function
new_df2.info( )
```

```python
# the .notnull() function gives a 'true' boolen value if there is no null value and a 'false' if there is a null value
# the .isnull() function gives a 'true' boolen value if there is a null value and a 'false' if there is no null value

pd.notnull( new_df2 )
```

```python
pd.isnull( new_df2 )
```

```python
# we can even pass it in a column
pd.isnull( new_df2[' year (2018)'])
```

```python
# we can also see the number of null or not null values as
pd.isnull( new_df2).sum( )
```

```python
pd.notnull( new_df2 ).sum( )
```

```python
# we can drop the rows with null values in the data frame as
new_df2.dropna( )
# as you can see it's only left out the rows with all columns containing values
```
